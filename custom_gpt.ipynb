Custom GPT app with langchain

```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)
```

```python
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import FileChatMessageHistory

llm = ChatOllama(model='mistral', temperature=1)

prompt = ChatPromptTemplate.from_messages([
    ("system", "you are a chatbot having a conversation with an intelligent human"),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{content}")
])

chain = prompt | llm
store = {}

def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = FileChatMessageHistory(
            file_path=f"chat_history_{session_id}.json"
        )
    return store[session_id]

chain_with_memory = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key='content',
    history_messages_key='chat_history',
)

while True:
    content = input("your prompt : ")
    if content.lower() in ['quit','exit','bye']:
        print("Goodbye!")
        print("-"*100)
        break
    response = chain_with_memory.invoke(
        {"content" : content}, 
        config={"configurable" : {"session_id" : "user_1"}} 
    )
    print(response.content)
    print("-"*100)
```
